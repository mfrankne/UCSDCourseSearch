```python
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
with open('data/schedule_of_classes.htm', 'r', encoding='utf-8') as file:
    html_content = file.read()
```

```python
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
tr_tags = soup.find_all('tr') # get all table tags
```

```python
all_codes = pd.read_csv('data/course_catalog.csv')
all_codes
```

```python
# Initialize a set to hold all course codes
course_codes = set()
code_lst = list(all_codes['Code'])
cur_dept = ''

for tag in tr_tags:
    # if it is a department tag
    h2_tag = tag.find('h2')
    if h2_tag and '(' in h2_tag.text:
        cur_dept = h2_tag.text.strip().split()[-2].replace('(', '')
    else:
        for td in tag.find_all(class_="crsheader"):
            course = cur_dept + ' ' + td.text
            if re.search('^[A-Z]{2,}\s\d{1,}[A-Z]?$', course) and cur_dept in all_codes['Department'].unique():
                course_codes.add(course)
```

## WebReg

```python
num_pages = 20
web_reg_courses = set()

for i in range(1, num_pages + 1):
    print("Working on page", i)
    file_path = "data/web_reg/sp24_web_reg_" + str(i) + ".html"
    
    # open the html file and read it in
    with open(file_path, 'r', encoding='utf-8') as file:
        web_reg_content = file.read()
        
    # parse the html file
    soup = BeautifulSoup(web_reg_content, 'html.parser')
    
    # get all the html tags that contain the course codes
    course_tags = soup.find_all(attrs={'id': 'search-group-header-id'}) # get all table tags
    
    # iterate through the tags and grab all the course codes and add to web_reg_courses
    for tag in course_tags:
        course = ' '.join(re.split('\s+', tag.find('tr').text.strip())[:2])
        web_reg_courses.add(course)
```
